### Train data 생성
- 부품이 붙어있지 않은 빈 PCB 이미지 위에 IC, capacitor, resistor 이미지를 배치함
- 부품의 크기와 위치는 random 함수를 이용하여 무작위로 설정
- 배치된 부품에 yolo 형식의 bounding box를 그린 후 txt 파일로 저장
- 총 29개의 학습데이터 생성

Train data 생성 시 PCB 기판 위에 배치한 개별 부품의 사이즈와 위치는 랜덤으로 무작위하게 설정하여 배치하였습니다.
따라서 재현 시 학습데이터와 가중치 파일이 바뀔 수 있습니다.

리더보드 기준 가장 높은 점수를 낸 가중치와 학습에 사용된 데이터는 아래 경로에 있습니다.
- 가중치 파일 : ./weights/best.pt
- 학습 데이터 : ./preprocessed_data/

```python
# IC

k = 0
step = 6

for j in range(10):
    part = "IC"

    path = "./data/train/" + part
    files = os.listdir(path)

    pcb_path = "./data/train/PCB/"
    pcb_files = os.listdir(pcb_path)

    pcb = cv2.imread("./data/train/PCB/" + pcb_files[j])

    file_name = "pcb_" + part
    txt = open("./data/train/labels/" + file_name + "_" + str(j) + ".txt", "w")
    for i in range(k,k+step):
        basewidth = random.randrange(50, 300)
        img = cv2.imread(path + "/" + files[i])

        wpercent = (basewidth/float(img.shape[0]))
        hsize = int((float(img.shape[1])*float(wpercent)))
        img = cv2.resize(img, (hsize, basewidth), interpolation = cv2.INTER_LANCZOS4)

        h, w = img.shape[:2]
        left = random.randrange(0, pcb.shape[0])
        right = random.randrange(0, pcb.shape[1])

        X = (right + w/2) / pcb.shape[1]
        y = (left + h/2) / pcb.shape[0]
        width = w / pcb.shape[1]
        height = h / pcb.shape[0]

        try:
            pcb[left:left+h, right:right+w] = img
            txt.write(f"0 {X} {y} {width} {height} \n")
        except:
            pass

    k += 6

    cv2.imwrite("./data/train/images/" + file_name + "_" + str(j) + ".jpg", pcb)
    txt.close()
```
