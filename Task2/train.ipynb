```python
# 라이브러리 import
import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split
```

### Train Test Split
- 학습데이터와 검증데이터를 0.75:0.25 비율로 분리

```python
local_img_path = "./data/train/images"
local_label_path = "./data/train/labels"
img_list = os.listdir(local_img_path)
label_list = os.listdir(local_label_path)
```

```python
# Train, Validation
image_train, image_valid, label_train, label_valid = train_test_split(img_list, label_list, test_size = 0.25, random_state = 42)
```

```python
for i in image_train:
    shutil.copy(local_img_path + "/" + i, "./data/train/train/images")
for i in image_valid:
    shutil.copy(local_img_path + "/" + i, "./data/train/valid/images")
for i in label_train:
    shutil.copy(local_label_path + "/" + i, "./data/train/train/labels")
for i in label_valid:
    shutil.copy(local_label_path + "/" + i, "./data/train/valid/labels")
```

### Train
- pre-trained 된 yolov5s 가중치를 이용하여 데이터 학습
