```python
# 라이브러리 import
import os
import natsort
import joblib
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
```

```python
cd /content/gdrive/MyDrive/AIFactory/Task1
```

## Train

```python
# 경로 설정
path = "./data/train_npy/"

files = natsort.natsorted(os.listdir(path))
```

```python
# Train dataset 생성
X_ans = []

for i in range(6000):
    df_ans = np.load(path + files[i])

    scaler = MinMaxScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df_ans))

    sum_ans = df_scaled.sum(axis = 1)

    X_ans.append(sum_ans)
```

```python
y_ans = pd.read_csv("./data/train.csv").label
```

```python
# 모델 생성 및 학습
model_xgb = XGBClassifier(random_state = 42)

model_xgb.fit(X_ans, y_ans)
```

## Test

```python
# 경로 설정
test_path = "./data/test_npy/"

test_files = natsort.natsorted(os.listdir(test_path))
```

```python
# Test dataset 생성
X_test = []

for j in range(3000):
    df_test = np.load(test_path + test_files[j])

    scaler = MinMaxScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df_test))

    sum_test = df_scaled.sum(axis = 1)

    X_test.append(sum_test)
```

```python
# predict 및 결과 파일 저장
y_ans_pred = model_xgb.predict(X_test)

answer = pd.read_csv("./data/anwer_sample.csv")

answer.label = y_ans_pred

answer.to_csv("./answer_test.csv", index = False)
```
